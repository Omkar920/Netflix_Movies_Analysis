Certificate Count:-

This code visualizes the distribution of movie certificates in the netflix_data dataset using a bar plot.


Key Step:-
Counts the certificate types using value_counts().


Creates a figure with plt.figure(figsize=(10, 6)) to set the plot size.


Plots a bar chart with orange-colored bars and black edges.


Labels the axes and title for clarity.


Rotates x-axis labels to improve readability.


Adds a grid for better visualization.


Displays the plot with plt.show().

This helps you quickly see how many movies fall under each certificate category.

```python
certificate_counts = netflix_data['certificate'].value_counts()
plt.figure(figsize=(10, 6))
certificate_counts.plot(kind='bar', color='orange', edgecolor='black')
plt.xlabel("Certificate")
plt.ylabel("Frequency")
plt.title("Bar Plot of Movie Certificates")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Genre Tend:-

Genre Frequency in movies dataset:-

This code creates a bar plot to visualize the top 10 most frequent genres in the netflix_data dataset.

Key Steps:
Counts occurrences of each genre using .value_counts().


Selects the top 10 genres with .nlargest(10).


Defines figure size using plt.figure(figsize=(9, 4)).


Plots a bar chart with genre_counts.plot(kind='bar'), coloring bars skyblue with black edges.


Labels the axes and sets a meaningful title.


Rotates x-axis labels for readability.


Adds grid lines to improve visualization.

This helps you quickly see the most common movie genres in your dataset.

```python
genre_counts =netflix_data['genre'].value_counts().nlargest(10)
plt.figure(figsize=(9, 4))
genre_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel("genre")
plt.ylabel("Frequency")
plt.title("genre Frequency in Movies Dataset")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Movie Ratings:-

This code creates a histogram to visualize the distribution of movie ratings in the netflix_data dataset.

Key Steps:
Defines figure size using plt.figure(figsize=(10, 6)).


Plots a histogram of ratings using plt.hist(), where:bins=20 divides the rating range into 20 intervals.color='purple' sets the bars to purple.
edgecolor='black' adds black borders to the bars for clarity.


Labels axes and sets a title for context.


Adds a grid along the y-axis to enhance readability.


Displays the plot with plt.show().


This gives a clear view of the rating distribution, helping analyze how movies are rated

```python
plt.figure(figsize=(10, 6))
plt.hist(netflix_data['rating'], bins=20, color='purple', edgecolor='black')
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.title("Histogram of Movie Ratings")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
